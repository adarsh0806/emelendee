## Welcome Node

```python
trace1 = go.Bar(
    y = chartio_1_plot['churned_trial_users'],
    x = chartio_1_plot['last seen position'],
    name='Churned Trial Users'
)
trace2 = go.Bar(
    y = chartio_1_plot['churned_paid_users'],
    x = chartio_1_plot['last seen position'],
    name='Churned Paid Users'
)
trace3 = go.Bar(
    y = chartio_1_plot['active_paid_users'],
    x = chartio_1_plot['last seen position'],
    name='Active Paid Users'
)

data = [trace1, trace2, trace3]

layout = go.Layout(
    barmode='group',
    xaxis=dict(
        autotick=False,
        ticks='inside',
        tick0=0,
        dtick=0.25,
        ticklen=8,
        tickwidth=4,
        tickcolor='#000'
    ),
    yaxis=dict(
        autotick=True,
        ticks='outside',
        tick0=0,
        dtick=0.25,
        ticklen=8,
        tickwidth=4,
        tickcolor='#000'
    ),
    autosize=False,
    width=800,
    height=700,
    margin=go.Margin(
        l=50,
        r=50,
        b=100,
        t=100,
        pad=4
    ),
    paper_bgcolor='#e5e5e5',
    plot_bgcolor='#c7c7c7'
)


fig = go.Figure(data=data, 
                layout=layout)
py.iplot(fig, 
         filename='grouped-bar')
```
